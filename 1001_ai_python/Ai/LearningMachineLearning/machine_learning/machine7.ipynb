```python
import pandas as pd
```

```python
dataset = pd.read_csv(r'C:\\Users\\Aliasger B\\1001_ai\\1001_ai_python\\Ai\\LearningMachineLearning\\datasets\\Social_Network_Ads.csv')
dataset
```

```python
x = dataset.iloc[:, [1,2,3]].values
feature_names = ['Gender', 'Age', 'EstimatedSalary']
```

```python
x
```

```python
y = dataset.iloc[:, -1].values
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
gender_label = LabelEncoder()
x[:, 0] = gender_label.fit_transform(x[:, 0])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test  = train_test_split(x, y, test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.naive_bayes import GaussianNB
Gaussian_nb = GaussianNB()

Gaussian_nb.fit(x_train, y_train)
```

```python
Gaussian_nb.predict(x_test)
```

```python
Gaussian_nb.score(x_test, y_test)
```

```python
from sklearn.inspection import permutation_importance

# Assuming you have already trained your Gaussian Naive Bayes model (e.g., nb_model)
result = permutation_importance(Gaussian_nb, x, y, n_repeats=30, random_state=0)

# Create a pandas Series for the importances
feature_imp = pd.Series(result.importances_mean, index=feature_names).sort_values(ascending=False)
```
