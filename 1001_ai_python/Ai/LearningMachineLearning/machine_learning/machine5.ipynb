```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
dataset= pd.read_csv(r'C:\\Users\\Burhanuddin\\1001_ai\\1001_ai_python\\Ai\\LearningMachineLearning\\datasets\\tennis.csv')
dataset
```

```python
from sklearn.preprocessing import LabelEncoder
Outlook = LabelEncoder()
Temperature = LabelEncoder()
Humidity = LabelEncoder()
Wind = LabelEncoder()
PlayTennis = LabelEncoder()
```

```python
dataset['Outlook'] = Outlook.fit_transform(dataset['Outlook'])
dataset['Temperature'] = Temperature.fit_transform(dataset['Temperature'])
dataset['Humidity'] = Humidity.fit_transform(dataset['Humidity'])
dataset['Wind'] = Wind.fit_transform(dataset['Wind'])
dataset['PlayTennis'] = PlayTennis.fit_transform(dataset['PlayTennis'])

dataset
```

```python
features_col = ['Outlook', 'Temperature', 'Humidity', 'Wind']
x = dataset[features_col]
y = dataset.PlayTennis
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='gini')
classifier.fit(x_train, y_train)
```

```python
classifier.predict(x_test)
```

```python
x_test
```

```python
y_test
```

```python
classifier.score(x_test, y_test)
```

```python
from sklearn import tree
tree.plot_tree(classifier)
```
