Merge Function

```python
import pandas as pd
csv36 = pd.DataFrame({"A" : [1,2,3,4,5,6], "B" : [7,8,9,10,11,12]})
csv37 = pd.DataFrame({"A" : [1,2,3,4,5,6], "C" : [13,14,15,16,17,18]})

# pd.merge(csv37, csv36, on='A')
pd.merge(csv36, csv37, on='A')
```

```python
import pandas as pd
csv38 = pd.DataFrame({"A" : [1,2,3,4,5,6], "B" : [7,8,9,10,11,12]})
csv39 = pd.DataFrame({"A" : [1,2,3,4,9,6], "C" : [13,14,15,16,17,18]})

pd.merge(csv38, csv39, on='A')
```

```python
import pandas as pd
csv40 = pd.DataFrame({"A" : [1,2,3,4,5,6], "B" : [7,8,9,10,11,12]})
csv41 = pd.DataFrame({"A" : [1,2,3,4,9,6], "C" : [13,14,15,16,17,18]})

# pd.merge(csv40, csv41, how= 'right')
pd.merge(csv40, csv41, how= 'outer', indicator= True)
# pd.merge(csv40, csv41, how= 'left')
```

```python
import pandas as pd
csv40 = pd.DataFrame({"A" : [1,2,3,4,5,6], "B" : [7,8,9,10,11,12]})
csv41 = pd.DataFrame({"A" : [1,2,3,4,9,6], "B" : [13,14,15,16,17,18]})

pd.merge(csv40, csv41, left_index=True, right_index=True, suffixes=("javascript", "python"))
```

Join

```python
import pandas as pd

join1 = pd.DataFrame({"A" : [1, 2, 3 ,4], "B" : [5, 7, 15, 16 ]}, index = ["a", "b","c", "d" ])
join2 = pd.DataFrame({"C" : [9, 10], "D" : [13 , 14]}, index = ["a", "b"])

join1.join(join2)
```

```python
join2.join(join1)
```

```python
# how parameter ensure that according to which list to show data 
join2.join(join1, how = "right")
```

```python
# outer = union , inner = intersection
join2.join(join1, how = "outer")
```

```python
import pandas as pd

join3 = pd.DataFrame({"A" : [1, 2, 3 ,4], "B" : [5, 7, 15, 16 ]}, index = ["a", "b","c", "d" ])
join4 = pd.DataFrame({"C" : [9, 10], "B" : [13 , 14]}, index = ["a", "b"])

# outer = union , inner = intersection
join4.join(join3, how = "inner", lsuffix="_left", rsuffix="_right")
```

pivot

```python
import pandas as pd

pivot1 = pd.DataFrame({"days" : [1, 2, 3, 4, 5, 6],
"st_name" : ["a", "b", "c", "a", "b", "c"],
"eng" : [13, 64, 75, 53, 75, 74], "math" : [32, 43 ,35 ,46 ,24 ,74]})
pivot1
```

```python
pivot1.pivot(index="days", columns="st_name")
```

```python
pivot1.pivot(index="days", columns="st_name", values="eng")
```

```python
import pandas as pd

pivot2 = pd.DataFrame({"days" : [1, 2, 1, 1, 1, 2],
"st_name" : ["a", "b", "c", "a", "b", "c"],
"eng" : [13, 64, 75, 53, 75, 74], "math" : [32, 43 ,35 ,46 ,24 ,74]})
pivot2.pivot_table(index="st_name", columns="days", aggfunc="mean")
```

```python
pivot2.pivot_table(index="st_name", columns="days", aggfunc="sum")
```

```python
pivot2.pivot_table(index="st_name", columns="days", aggfunc="mean", margins=True)
```
